```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

plt.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# for Palatino and other serif fonts use:
plt.rc('font',**{'family':'serif','serif':['Palatino']})
plt.rc('text', usetex=False)

%matplotlib inline
```

```python
## Problem 1 Plot a)
x = np.arange(-5,5, 0.1)
y = 1 +3*x
y_top = np.full((len(x),), 16)
y_bottom = np.full((len(x),), -14)
fig, ax = plt.subplots(1,1, figsize=(12,12))

ax.plot(x, y)
ax.fill_between(x, y, y_top, color='red', alpha=0.1)
ax.fill_between(x, y, y_bottom, color='blue', alpha=0.1)

ax.set_xlabel(r'X_1')
ax.set_ylabel(r'X_2')
ax.set_title(r'Hyperplane split')

ax.text(s='>0', x=2, y=-5)
ax.text(s='<0', x=-2, y=5)

plt.savefig('q1_pa_hyperplane.png')

```
