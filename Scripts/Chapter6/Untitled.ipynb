```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
## Ridge regression penalized RSS
y = 5
lam = 5
beta = np.arange(0, 100, 0.5)
prss = (lam + 1)*beta**2 - 2*beta*y + y**2
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(beta, prss)
ax1.set_xlim(-1,4)
ax1.set_ylim(0,50)

ax1.set_title("Ridge Regression PRSS")
ax1.set_xlabel("Beta")
ax1.set_ylabel("PRSS")

fig.savefig("Ridge_PRSS_toy_model.png")
```
