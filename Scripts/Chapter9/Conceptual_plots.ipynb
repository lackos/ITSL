```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi
import matplotlib.patches as mpatches

plt.rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# for Palatino and other serif fonts use:
plt.rc('font',**{'family':'serif','serif':['Palatino']})
plt.rc('text', usetex=False)

%matplotlib inline
```

```python
## Problem 1 Plot a)
x = np.arange(-5,5, 0.1)
y = 1 +3*x
y_top = np.full((len(x),), 16)
y_bottom = np.full((len(x),), -14)
fig, ax = plt.subplots(1,1, figsize=(12,12))

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.plot(x, y)
ax.fill_between(x, y, y_top, color='red', alpha=0.1)
ax.fill_between(x, y, y_bottom, color='blue', alpha=0.1)

ax.set_xlabel(r'X_1')
ax.set_ylabel(r'X_2')
ax.set_title(r'H_1 Hyperplane split')

ax.set_xlim(-5,4.8)
ax.set_ylim(-14,16)

ax.text(s='>0', x=2, y=-5)
ax.text(s='<0', x=-2, y=5)


plt.savefig('q1_pa_hyperplane.png')

```

```python
## Problem 1 Plot b)
x = np.arange(-5,5, 0.1)
y = 1 -0.5*x
y_top = np.full((len(x),), 16)
y_bottom = np.full((len(x),), -14)
fig, ax = plt.subplots(1,1, figsize=(12,12))

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.plot(x, y)
ax.fill_between(x, y, y_top, color='blue', alpha=0.1)
ax.fill_between(x, y, y_bottom, color='red', alpha=0.1)

ax.set_xlabel(r'X_1')
ax.set_ylabel(r'X_2')
ax.set_title(r'H_2 Hyperplane split')

ax.set_xlim(-5,4.8)
ax.set_ylim(-14,16)

ax.text(s='<0', x=2, y=-5)
ax.text(s='>0', x=-2, y=5)

plt.savefig('q1_pb_hyperplane.png')
```

```python
## Problem 2 Plot a)


u=-1     #x-position of the center
v=2    #y-position of the center
a=2     #radius on the x-axis
b=2    #radius on the y-axis

t = np.linspace(0, 2*pi, 100)
x = u+a*np.cos(t)
y = v+b*np.sin(t)

circle = mpatches.Circle((-1,2), radius=2, facecolor='r', alpha=0.1, edgecolor=None)


fig, ax = plt.subplots(1,1, figsize=(12,12))

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.plot(x, y, color='k')
ax.add_patch(circle)

ax.set_xlim(-4,2)
ax.set_ylim(-1,5)

ax.set_xlabel(r'X_1')
ax.set_ylabel(r'X_2')
ax.set_title(r'Non-linear Hyperplane')

ax.text(s='<0', x=-1, y=2)
plt.savefig('q2_hyperplane.png')
plt.show()
```

```python
## Problem 3 Plot a)
x_r = [3,2,4,1]
y_r = [4,2,4,4]

x_b = [2,4,4]
y_b = [1,3,1]

x = np.arange(0.5,5, 0.1)
y_h = x - 0.5
y_u = x
y_l = x - 1

fig, ax = plt.subplots(1,1, figsize=(12,12))

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.scatter(x_r, y_r, color='r', s=100)
ax.scatter(x_b, y_b, color='b', s=100)

ax.plot(x,y_h, color='k')
ax.plot(x,y_u, color='k', ls='--')
ax.plot(x,y_l, color='k', ls='--')

ax.fill_between(x, y_h, y_u, color='black', alpha=0.1)
ax.fill_between(x, y_h, y_l, color='black', alpha=0.1)

ax.set_xlabel(r'X_1')
ax.set_ylabel(r'X_2')
ax.set_title(r'Toy Data set')

plt.savefig('q3_pd_hyperplane.png')
```
