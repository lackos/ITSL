```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Problem 1
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,12))

ax.grid(True, which='both')
# ax.axhline(y=0, color='k')
# ax.axvline(x=0, color='k')

ax.hlines(y=-50, xmin=0.0, xmax=150.0, color='b',lw=3)
ax.hlines(y=0, xmin=250, xmax=500, color='b',lw=3)

ax.vlines(x=150, ymin=-100, ymax=100.0, color='r',lw=3)
ax.vlines(x=250, ymin=-100, ymax=100, color='r',lw=3)
ax.vlines(x=400, ymin=0, ymax=100, color='r',lw=3)

ax.text(s='R_1', x=75, y=-75, fontsize='x-large')
ax.text(s='R_2', x=75, y=25, fontsize='x-large')
ax.text(s='R_3', x=200, y=0, fontsize='x-large')
ax.text(s='R_4', x=387.5, y=-50, fontsize='x-large')
ax.text(s='R_5', x=312.5, y=50, fontsize='x-large')
ax.text(s='R_6', x=450, y=50, fontsize='x-large')

ax.set_xlim(0, 500)
ax.set_ylim(-100,100)

ax.set_xlabel(r'X_1', fontsize='x-large')
ax.set_ylabel(r'X_2', fontsize='x-large')
ax.set_title('Example Tree partition of a 2D feature space', fontsize='xx-large')

plt.savefig('q1_tree_partition.png')
```
