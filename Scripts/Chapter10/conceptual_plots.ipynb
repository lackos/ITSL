```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12,12))

x_r = [1,0,6]
y_r = [4,4,2]

x_b = [1, 5, 4]
y_b = [3, 1, 0]

xr_cluster = [2.333]
yr_cluster = [3.333]

xb_cluster = [3.3333]
yb_cluster = [1.3333]

ax.grid(True, which='both')
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

ax.scatter(x_r, y_r, color='r', s=100)
ax.scatter(x_b, y_b, color='b', s=100)

ax.scatter(xr_cluster, yr_cluster, color='r', s=250, marker='x')
ax.scatter(xb_cluster, yb_cluster, color='b', s=250, marker='x')


ax.set_xlabel(r'X_1')
ax.set_ylabel(r'X_2')
ax.set_title(r'Kmeans clustering toy dataset (Initial)')

plt.savefig("p10_parta_clusters.png", dpi=500)
```
