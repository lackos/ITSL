```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
## Problem 2 part g

x = np.arange(1, 10000000, 1)
y = 1 - ((x-1)/x)**x

fig, ((ax1), (ax2)) = plt.subplots(nrows=2, ncols=1, figsize=(12,8))

ax1.plot(x, y)
ax1.set_ylabel("p(x)")
ax1.set_xlabel("n")

ax2.plot(x, y)
ax2.set_xscale('log')
ax2.set_ylabel("p(x)")
ax2.set_xlabel("n")

plt.suptitle("Probability that some observation lies in a bootstrap sample for sample size n")

fig.tight_layout(rect=[0, 0.00, 1, 0.95])

plt.savefig("bootstrap_limit.png", dpi=500)
```

```python
## Problem 2 part h
import random

## Generate orginal sample set
orig = np.arange(0, 100, 1)

## Assign jth element (here is it the 4th element)
test_obs = orig[3]

## define number of bootstraps to perform
num_boots = 20

## Create list to score whether the test obs was in the bootstrap sample
scores = []

## Iterate over the number of bootstraps
for j in range(num_boots):
    ## Create the bootstrap
    for i in range(100):
        ## Randomly select a point from the original dataset
        rand_obs = random.choice(orig)
        ## If if the test obs is selected score bootstrap true and break the loop
        if rand_obs == test_obs:
            scores.append(True)
            break
    ## if the test obs wasn't selected score the bootstrap false.
    scores.append(False)
    
prob = sum(scores)/len(scores)

# print(scores)
print("Number of times the observation was in the bootstrap samples: ", sum(scores))
print("Probability: ", prob)
```
